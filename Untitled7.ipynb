```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import distance_transform_edt
import os

# Load and preprocess the image
img = cv2.imread('image.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize if necessary
h, w, c = img.shape
if w > 1000:
    new_w = 1000
    ar = w / h
    new_h = int(new_w / ar)
    img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)

# Adaptive thresholding for uneven lighting
def adaptive_thresholding(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    adaptive_thresh = cv2.adaptiveThreshold(
        img_gray,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY_INV,
        11,
        2,
    )
    return adaptive_thresh

thresh_img = adaptive_thresholding(img)

# Morphological closing to connect broken parts
kernel = np.ones((3, 3), np.uint8)
closed = cv2.morphologyEx(thresh_img, cv2.MORPH_CLOSE, kernel, iterations=1)

# Distance Transform
dist_transform = distance_transform_edt(closed)
dist_transform = cv2.normalize(dist_transform, None, 0, 1.0, cv2.NORM_MINMAX)

# Threshold to find sure foreground
_, sure_fg = cv2.threshold(dist_transform, 0.5, 1.0, cv2.THRESH_BINARY)
sure_fg = (sure_fg * 255).astype(np.uint8)

# Dilate to get sure background
sure_bg = cv2.dilate(closed, kernel, iterations=3)

# Find unknown region
unknown = cv2.subtract(sure_bg, sure_fg)

# Marker labeling
_, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1  # Increment to ensure background is not zero
markers[unknown == 255] = 0

# Apply watershed
img_markers = img.copy()
markers = cv2.watershed(img_markers, markers)
img_markers[markers == -1] = [255, 0, 0]  # Mark boundaries in red

# Extract characters with bounding boxes
character_regions = []
for marker in np.unique(markers):
    if marker <= 1:  # Ignore background and boundary
        continue

    # Create mask for each marker
    mask = np.zeros(closed.shape, dtype="uint8")
    mask[markers == marker] = 255

    # Find contours for the current marker
    cnts, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in cnts:
        x, y, w, h = cv2.boundingRect(cnt)
        if w < 5 or h < 5:  # Filter very small regions
            continue
        character_regions.append((x, y, w, h))

# Sort characters by their position (left-to-right, top-to-bottom)
character_regions = sorted(character_regions, key=lambda b: (b[1], b[0]))  # Sort by y, then x

# Save characters to separate files
output_folder = "extracted_characters"
os.makedirs(output_folder, exist_ok=True)  # Create folder if not exists

for i, (x, y, w, h) in enumerate(character_regions):
    char_img = closed[y:y+h, x:x+w]
    file_path = os.path.join(output_folder, f"character_{i+1}.png")
    cv2.imwrite(file_path, char_img)
    print(f"Saved: {file_path}")

# Display result with bounding boxes
for x, y, w, h in character_regions:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 1)

plt.figure(figsize=(10, 6))
plt.title("Segmented Characters with Bounding Boxes")
plt.imshow(img)
plt.axis('off')
plt.show()
```
